<h2 Style="color:green">Pregunta No 3<h2>
<p>¿En cuál ciudad o pueblo se dan mayor cantidad de ventas de propiedades? Y ¿Qué elementos particulares tienen esas ciudades o pueblos?</p>

```python
mpl.rcParams["axes.spines.bottom"]=True
mpl.rcParams["axes.spines.left"]=True
mpl.rcParams["axes.spines.right"]=False
mpl.rcParams["axes.spines.top"]=False
```

```python
gf=dfVenta.groupby (['town']).count()
gf=gf.iloc[:,:1]
gf=gf.sort_values('serialNumber', ascending=False)
gf=gf.head(10)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,6.9), dpi=100)
 
# Datos a graficar
town = gf.index.tolist()
y_pos = np.arange(len(town))
 
cant = gf['serialNumber'].tolist()
 
barColors = ['#ebd07f','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5']
 
# Generacion del grafico
ax.barh(y_pos, cant, align='center', color = barColors)
ax.set_yticks(y_pos, labels=town)
ax.invert_yaxis()  # labels read top-to-bottom

#formato del titulo
 
title = "Grafico 1: Top 10 de pueblos con mayor venta de propiedades"
plt.suptitle(title,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del subtitulo
stitle = "Pueblo"
ax.set_title(stitle,
            horizontalalignment = 'left',
            x = 0.08,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 12,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del eje X
etiqueta = "Cantidad de propiedades"
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = "bold",
             fontfamily = "serif")
 
#Configuracion de coordenadas de la etique del eje x
ax.xaxis.set_label_coords(x=0, y= -0.05, transform= ax.transAxes)
 
#configuracion del tamano
ax.tick_params(axis='both', 
               labelsize = 10)
 
#Congiruacion de la ubicacion del grafico
plt.subplots_adjust(top=0.80)
 
 
plt.show()
```

<h2 Style="color:green">Pregunta No 1<h2>
<p>¿Cuál es el ingreso anual de las propiedades vendidas?</p>

```python
añoGroup=dfVenta[['listYear','saleAmount']]
añoGroup.head(10)
```

```python
ga=añoGroup.groupby(['listYear'])['saleAmount'].sum().reset_index(name ='prom')
ga=ga.sort_values('prom',ascending=False)
ga=ga.head(10)
ga
```

```python
fig, ax = plt.subplots(1,1, figsize = (15,6.9), dpi=100)
 
# Datos a graficar
prom = ga.index.tolist()
y_pos = ga.listYear
y_pos = np.arange(len(prom))
cant = ga['prom'].tolist()
 

# Generacion del grafico
ax.barh(y_pos, cant, align='center')
ax.set_yticks(y_pos, ga.listYear)
ax.invert_yaxis()  # labels read top-to-bottom

#formato del titulo
 
title = "Grafico 5: Top 10 Ingresos por año de las propiedades vendidas"
plt.suptitle(title,
            horizontalalignment = 'left',
            x = 0.001,
            y = 0.87,
            transform = fig.transFigure,
            fontsize = 20,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del subtitulo
stitle = "Año"
ax.set_title(stitle,
            horizontalalignment = 'left',
            x = 0.08,
            y = 0.80,
            transform = fig.transFigure,
            fontsize = 12,
            fontweight = "bold",
            fontfamily = "serif")
 
#formato del eje X
etiqueta = "1 por cada 10 mil millones"
ax.set_xlabel(etiqueta,
             ha = 'left',
             fontsize = 10,
             fontweight = "bold",
             fontfamily = "serif")
 
#Configuracion de coordenadas de la etique del eje x
ax.xaxis.set_label_coords(x=0, y= -0.05, transform= ax.transAxes)
 
#configuracion del tamano
ax.tick_params(axis='both', 
               labelsize = 10)
 
#Congiruacion de la ubicacion del grafico
plt.subplots_adjust(top=0.80)
 
 
plt.show()
```
